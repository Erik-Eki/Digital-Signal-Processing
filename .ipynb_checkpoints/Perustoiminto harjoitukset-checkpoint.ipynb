## Tehtävä 1

```octave
function taulukko(rivi, sarake)

    x = ones(rivi,sarake);
    for i=1:rivi
        for j=1:sarake
            x(i,j)=i+j;
        endfor
    endfor
    disp(x)
    
endfunction

taulukko(3,4);
```

## Tehtävä 3

```octave
function sinsignal(A, f, theta, pituus)
    fs = 20*f; # näytteistystaajuus
    dt = 1/fs; # kahden näytepisteen välinen etäisyys
    t = 0;
    sini = 0;
    npiste = fs*pituus;
    
    rad = deg2rad(theta); # asteet radiaaneiksi
    t = 0:dt:pituus-dt; # aikavektori aika-askeleella dt
    
    sini=A*sin(2*pi*f*t+rad); # sinisignaali

    subplot(3,1,1)
    plot(t,sini)
    subplot(3,1,2)
    plot(t)
    subplot(3,1,3)
    stem(t)
    
endfunction

sinsignal(10,5,3,1);
```

## Tehtävä 4

```octave
function NoiseSine(A, f, theta, pituus)
    fs = 50*f; # näytteistystaajuus
    dt = 1/fs; # kahden näytepisteen välinen etäisyys
    t = 0;
    sini = 0;
    npiste = fs*pituus;
    
    rad = deg2rad(theta); # asteet radiaaneiksi
    t = 0:dt:pituus-dt; # aikavektori aika-askeleella dt
    
    sini=A*sin(2*pi*f*t+rad); # sinisignaali
    
    kohina = rand(1,length(sini)); # kohina
    max_kohina = max(kohina); # max kohina
    disp(max_kohina)
    skaalaus = (A/max_kohina)*0.1; # skaalauskerroin 
    disp(skaalaus)
    
    kohina_skaalattu = skaalaus*kohina;
    kohinasini = kohina_skaalattu+sini;

    subplot(3,1,1)
    plot(t,sini)
    title ("Sini");
    
    subplot(3,1,2)
    plot(t, kohina, 'g')
    title ("Kohina");
    
    subplot(3,1,3)
    plot(t, kohinasini, 'r')
    title ("Kohinasini");
    xlabel('Aika [s]')
    ylabel('Signaalin arvo')

    
endfunction

NoiseSine(10,5,3,1);
```
