```octave
function [] = sinsignal(A,f,theta,pituus,fs)

rad = deg2rad(theta); # asteet radiaaneiksi

if (fs < 2*f)
    printf("Näytteistajuus liian pieni: Signaali laskostuu");
    t1=0:1/fs:pituus-(1/fs);
    t2=0:1/(f*100):pituus-(1/(f*100));
    
    sini1=A*sin(2*pi*f*t1+rad);
    sini2=A*sin(2*pi*f*t2+rad);
    
    subplot(2,1,1)
    plot(t1,sini1, "b")
    title('Laskostunut sini')
    xlabel('Aika');ylabel('Signaalin arvo')
    subplot(2,1,2)
    title('Laskostumaton sini')
    plot(t2,sini2, "r")
    title('Laskostumaton sini')
    xlabel('Aika');ylabel('Signaalin arvo')

elseif (fs == 2*f)
    printf("Näytteistajuus tasan 2 kertaa signaalin taajuus. Näytteistajuutta kannattaa kasvattaa.");
    t1=0:1/fs:pituus-(1/fs);
    t2=0:1/(f*100):pituus-(1/(f*100));
    
    sini1=A*sin(2*pi*f*t1+rad);
    sini2=A*sin(2*pi*f*t2+rad);

    subplot(2,1,1)
    plot(t1,sini1, "b")
    title('Laskostumisen rajalla oleva sini')
    xlabel('Aika');ylabel('Signaalin arvo')
    subplot(2,1,2)
    title('Laskostumaton sini')
    plot(t2,sini2, "r")
    title('Laskostumaton sini')
    xlabel('Aika');ylabel('Signaalin arvo')

elseif (fs > 2*f)
    printf("Signaali ei laskostu");
    t=0:1/fs:pituus-(1/fs);
    sini=A*sin(2*pi*f*t+rad);
    
    subplot(2,1,1)
    plot(t,sini)
    xlabel('Aika');ylabel('Signaalin arvo')
    subplot(2,1,2)
    stem(t,sini, "g")
    xlabel('Aika');ylabel('Signaalin arvo')
endif

end
```

```octave
sinsignal(10,5,3,1,8);
```
